```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
books = soup.find_all('tr', class_ = 'book')
```

```python
isbn = []
title = []
author = []
language = []
pages = []
topic = []
price = []

for book in books:
    row = book.find_all('td')
    
    isbn.append(row[0].text)
    title.append(row[1].text)
    author.append(row[2].text)
    language.append(row[3].text)
    pages.append(row[4].text)
    topic.append(row[5].text)
    price.append(row[6].text)
    
book_dict = {
    'ISBN': isbn,
    'Title': title,
    'Author': author,
    'Language': language,
    'Pages': pages,
    'Topic': topic,
    'Price': price,
}

df = pd.DataFrame(book_dict)

df.head()
```

 Question 3 1 pts

Which topic of books is the most expensive, on average?


```python
# Remove non-numeric characters from 'Price' column
df['Price'] = df['Price'].replace('[\$,]', '', regex=True)

# Convert 'Price' column to float type
df['Price'] = df['Price'].astype(float)

df[['Topic', 'Price']].groupby('Topic').mean().sort_values(by='Price', ascending=False)
```

 Question 4 1 pts

Which topic of book has the most pages, on average?


```python
# Convert 'Price' column to integer type
df['Pages'] = df['Pages'].astype(int)

df[['Topic', 'Pages']].groupby('Topic').mean().sort_values(by='Pages', ascending=False)
```
